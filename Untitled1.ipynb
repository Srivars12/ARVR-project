```python
# Cell 2
import os
import sys
from glob import glob
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
import tensorflow as tf

# USER CONFIG — adjust if needed
DATASET_DIR = "Dataset"   # folder containing class subfolders & captions.csv
CAPTIONS_CSV = os.path.join(DATASET_DIR, "captions.csv")
IMAGE_SIZE = (224, 224)
BATCH_SIZE = 16
EPOCHS = 20
MODEL_OUT = "temple_object_model.h5"
TFLITE_OUT = "temple_object_model.tflite"
RANDOM_SEED = 42

print("TensorFlow version:", tf.__version__)
```

```python
import os
import pandas as pd

# Set your dataset paths
DATASET_DIR = r"C:\Users\reshm\OneDrive\Desktop\temple\images"
CAPTIONS_CSV = r"C:\Users\reshm\OneDrive\Desktop\temple\captions.csv"

# Check dataset structure
assert os.path.exists(DATASET_DIR), f"{DATASET_DIR} not found!"
print("Top-level files/folders:", os.listdir(DATASET_DIR))

# Load captions
captions_df = pd.read_csv(CAPTIONS_CSV)
print("✅ Captions loaded successfully!")
print("CSV Columns:", captions_df.columns.tolist())

# Rename to match standard variable name
captions_df.rename(columns={'image_path': 'imageName', 'captions': 'caption'}, inplace=True)

# Clean image path column
captions_df['imageName'] = captions_df['imageName'].astype(str).str.replace("\\\\", "/", regex=False).str.strip()

# Show sample data
captions_df.head()

```

```python
# ========================================
# Cell 4: Load images and labels (debug-friendly)
# ========================================
def gather_images_and_labels(dataset_dir, image_size):
    classes = [d for d in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, d))]
    classes.sort()
    print("Classes detected:", classes)
    class_to_index = {c:i for i,c in enumerate(classes)}

    images, labels, rel_paths = [], [], []

    for cls in classes:
        cls_dir = os.path.join(dataset_dir, cls)
        class_count = 0
        for ext in ("*.jpg","*.jpeg","*.png","*.bmp"):
            for p in glob(os.path.join(cls_dir, ext)):
                img = cv2.imread(p)
                if img is None:
                    print("Warning: can't read", p)
                    continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                img = cv2.resize(img, image_size)
                img = img.astype("float32") / 255.0
                images.append(img)
                labels.append(class_to_index[cls])
                rel_paths.append(os.path.relpath(p, dataset_dir).replace("\\","/"))
                class_count += 1
        print(f"{cls}: {class_count} images loaded")
        if class_count == 0:
            print(f"⚠️ Warning: no images found in folder {cls_dir}")

    total = len(images)
    print(f"\n✅ Total images loaded: {total}")
    return np.array(images), np.array(labels), np.array(rel_paths), classes

X, y, rel_paths, classes = gather_images_and_labels(DATASET_DIR, IMAGE_SIZE)
```

```python
# Cell 5
X_train, X_val, y_train, y_val, rp_train, rp_val = train_test_split(
    X, y, rel_paths, test_size=0.2, random_state=RANDOM_SEED, stratify=y
)
print("Train:", len(X_train), "Val:", len(X_val))
```

```python
# ========================================
# Cell 6: Build transfer learning model
# ========================================
def build_model(num_classes, image_size):
    base = tf.keras.applications.MobileNetV2(
        input_shape=(image_size[0], image_size[1], 3),
        include_top=False,
        weights="imagenet"
    )
    base.trainable = False
    inputs = tf.keras.Input(shape=(image_size[0], image_size[1], 3))
    x = base(inputs, training=False)
    x = tf.keras.layers.GlobalAveragePooling2D()(x)
    x = tf.keras.layers.Dense(512, activation="relu")(x)
    x = tf.keras.layers.Dropout(0.3)(x)
    outputs = tf.keras.layers.Dense(num_classes, activation="softmax")(x)
    model = tf.keras.Model(inputs, outputs)
    model.compile(optimizer=tf.keras.optimizers.Adam(1e-4),
                  loss="sparse_categorical_crossentropy",
                  metrics=["accuracy"])
    return model

model = build_model(len(classes), IMAGE_SIZE)
model.summary()
```

```python
# ========================================
# Cell 7: Train the model
# ========================================
callbacks = [
    tf.keras.callbacks.ModelCheckpoint(MODEL_OUT, save_best_only=True, monitor="val_accuracy", mode="max"),
    tf.keras.callbacks.EarlyStopping(monitor="val_accuracy", patience=6, restore_best_weights=True)
]

history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    callbacks=callbacks
)
```

```python
# ========================================
# Cell 8: Plot training curves
# ========================================
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='train loss')
plt.plot(history.history['val_loss'], label='val loss')
plt.legend(); plt.title("Loss")
plt.subplot(1,2,2)
plt.plot(history.history['accuracy'], label='train acc')
plt.plot(history.history['val_accuracy'], label='val acc')
plt.legend(); plt.title("Accuracy")
plt.show()
```

```python
# Cell 9
# Save Keras model (best model should already be saved by checkpoint)
model.save(MODEL_OUT)
print("Saved model:", MODEL_OUT)

# Convert to TFLite
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
with open(TFLITE_OUT, "wb") as f:
    f.write(tflite_model)
print("Saved TFLite:", TFLITE_OUT)
```

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Helper: Get caption for a relative path
def get_caption_for_relpath(captions_df, rel_path):
    """
    Match caption based on image filename (ignores folder path)
    """
    filename = os.path.basename(rel_path)
    m = captions_df[captions_df['imageName'].str.endswith(filename)]
    if not m.empty:
        return m['caption'].iloc[0]
    return "No caption found."

# Main prediction & display function
def predict_and_show(model, image_path, dataset_dir=DATASET_DIR):
    """
    Predict the class of an image and display it with its caption.

    Args:
        model: Trained Keras/TensorFlow model
        image_path: Absolute or relative path to image
        dataset_dir: Base dataset directory
    """
    # Convert to absolute path if relative
    if not os.path.isabs(image_path):
        image_path = os.path.join(dataset_dir, image_path)

    # Check if file exists
    if not os.path.exists(image_path):
        print("⚠️ Image file does not exist:", image_path)
        return

    # Read image
    img = cv2.imread(image_path)
    if img is None:
        print("⚠️ Unable to read image:", image_path)
        return

    # Convert BGR to RGB and resize
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img_resized = cv2.resize(img_rgb, IMAGE_SIZE).astype("float32") / 255.0

    # Predict class
    probs = model.predict(np.expand_dims(img_resized, 0))[0]
    idx = int(np.argmax(probs))
    cls = classes[idx]
    conf = probs[idx]

    # Get caption
    rel = os.path.relpath(image_path, dataset_dir).replace("\\","/")
    caption = get_caption_for_relpath(captions_df, rel)

    # Display results
    print(f"Predicted: {cls} (confidence {conf:.3f})")
    print("Caption:", caption)
    plt.figure(figsize=(6,6))
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(f"{cls} ({conf:.2f})\n{caption}")
    plt.show()
```
